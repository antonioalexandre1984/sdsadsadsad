```python
# 1.0 - Na coluna co_cid_causa_morte separa as cids por virgula.
# Carregamento do arquivo CSV: Lê um arquivo CSV contendo dados de óbitos e armazena em um DataFrame do Pandas. Utiliza a codificação latin1 e lida com linhas problemáticas exibindo um aviso.Verificação da existência da coluna: Confirma se a coluna co_cid_causa_morte está presente no DataFrame. Caso contrário, exibe uma mensagem de erro e interrompe a execução.Separação dos códigos CID: Define uma função chamada separate_cids que usa uma expressão regular para identificar e extrair códigos CID (padrão de uma letra seguida de três dígitos, por exemplo, 'A123') da coluna co_cid_causa_morte. Os códigos CID encontrados são separados por vírgulas e retornados como uma string única.Aplicação da função: Aplica a função separate_cids à coluna co_cid_causa_morte, armazenando o resultado na mesma coluna e em uma nova coluna chamada co_cid_causa_morte_a.Exibição dos resultados: Exibe o conteúdo das colunas co_cid_causa_morte e co_cid_causa_morte_a após a modificação.Salvamento do DataFrame atualizado: Salva o DataFrame modificado em um novo arquivo CSV, localizado no caminho especificado. Exibe uma mensagem de sucesso ou erro dependendo do resultado do salvamento.Em resumo, este código limpa e padroniza os códigos CID em uma coluna específica do DataFrame, separando múltiplos códigos em strings únicas, antes de salvar os resultados em um novo arquivo CSV.

import pandas as pd
import re

# Definir o caminho do arquivo CSV
file_path = r'C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Dev\dados_atualizados.csv'

# Carregar o arquivo CSV
try:
    df = pd.read_csv(file_path, delimiter=',', encoding='latin1', on_bad_lines='warn')  # Ajustar delimitador e codificação conforme necessário
except Exception as e:
    print(f"Erro ao carregar o arquivo CSV: {e}")
    exit()

df = df.reset_index(drop=True)

# Verificar se a coluna 'co_cid_causa_morte' existe
if 'co_cid_causa_morte' not in df.columns:
    print("A coluna 'co_cid_causa_morte' não foi encontrada no DataFrame.")
    exit()

# Função para separar CIDs agrupados e mantê-los como string única
def separate_cids(cids):
    if pd.isna(cids):
        return ''  # Retorna uma string vazia se o valor for NaN
    
    # Expressão regular para identificar padrões de CID
    cid_pattern = re.compile(r'[A-Z]\d{3}')
    # Encontrar todos os CIDs na string
    found_cids = cid_pattern.findall(str(cids))
    # Juntar os CIDs encontrados com vírgula
    return ','.join(found_cids)

# Aplicar a função à coluna 'co_cid_causa_morte'
df['co_cid_causa_morte'] = df['co_cid_causa_morte'].astype(str).apply(separate_cids)
df['co_cid_causa_morte_a'] = df['co_cid_causa_morte'].astype(str).apply(separate_cids)

# Verificar o DataFrame modificado
#print(df.head())
#print(df.columns)
print(df['co_cid_causa_morte'])
print(df['co_cid_causa_morte_a'])
print(df[['co_cid_causa_morte','co_cid_causa_morte_a']].dtypes)

# Definir o caminho para salvar o arquivo modificado
output_file_path = r'C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Teste\01-dados_atualizados.csv'

# Salvar o DataFrame modificado em um novo arquivo CSV
try:
    df.to_csv(output_file_path, index=False)
    print(f"Arquivo CSV modificado salvo em: {output_file_path}")
except Exception as e:
    print(f"Erro ao salvar o arquivo CSV: {e}")
```

```python
#11 - Este código realiza o processamento e atualização de dados de dois arquivos CSV: um contendo dados municipais e outro com informações de população. O objetivo é integrar informações da população no arquivo de dados municipais e exportar o resultado atualizado. O código segue as seguintes etapas:

#Definição dos Caminhos dos Arquivos

#csv_dados_path: Caminho para o arquivo CSV que contém os dados municipais.
#csv_populacao_utf8_path: Caminho para o arquivo CSV de população já convertido para UTF-8.

#Carregamento dos Dados

#Arquivo de Dados Municipais: Carrega o CSV contendo os dados municipais em um DataFrame (df_dados) usando a codificação UTF-8.

#Arquivo de População: Tenta carregar o CSV de população em um DataFrame (df_populacao) com as colunas ['COD_MUN', 'Macro', 'GRS', 'RS'], também usando a codificação UTF-8. Se a carga falhar, o código imprime um erro e encerra a execução.

#Preparação dos Dados

#Conversão de Tipos: Converte as colunas co_municipio_ibge do DataFrame df_dados e COD_MUN do DataFrame df_populacao para o tipo string, assegurando a compatibilidade para comparações subsequentes.

#Inicialização das Colunas: Adiciona três novas colunas ao DataFrame df_dados: pb_grs, pb_regiao_saude_new, e pb_macro, todas inicialmente definidas como None.
#Atualização das Colunas

#Preenchimento: Itera sobre cada linha do DataFrame df_dados. Para cada município (co_municipio_ibge), busca o código correspondente no DataFrame df_populacao. Se uma correspondência for encontrada, preenche as colunas pb_grs, pb_regiao_saude_new, e pb_macro com os valores de GRS, RS, e Macro, respectivamente.
#Exportação dos Dados Atualizados

#Salvamento: Salva o DataFrame atualizado (df_dados) em um novo arquivo CSV (dados_atualizados_com_grs_rs_macro2.csv) usando a codificação UTF-8.

#Verificação: Exibe os primeiros 20 registros do DataFrame atualizado para verificar a integridade e a precisão dos dados.

#Esse código proporciona uma forma eficiente de integrar dados de diferentes fontes e garantir que a codificação dos caracteres especiais seja preservada corretamente ao longo do processo.

import pandas as pd

# Caminhos dos arquivos
csv_dados_path = r"C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Dev\10-dados_atualizados_co_municipio_ibge.csv"
csv_populacao_path = r"C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Dev\TesteObitosSeparados.csv"

# Carregar o DataFrame de dados
df_dados = pd.read_csv(csv_dados_path, delimiter=',', encoding='utf-8')

# Carregar o arquivo CSV de população já convertido para UTF-8
try:
    df_populacao = pd.read_csv(csv_populacao_path, delimiter=',', encoding='utf-8', usecols=['COD_MUN', 'Macro', 'GRS', 'RS'])
    print("Arquivo CSV de população carregado com sucesso.")
except Exception as e:
    print(f"Erro ao carregar o arquivo CSV de população: {e}")
    exit()

# Garantir que as colunas 'co_municipio_ibge' e 'COD_MUN' sejam do tipo string
df_dados['co_municipio_ibge'] = df_dados['co_municipio_ibge'].astype(str)
df_populacao['COD_MUN'] = df_populacao['COD_MUN'].astype(str)

# Inicializar as colunas de destino no DataFrame de dados
df_dados['pb_grs'] = None
df_dados['pb_regiao_saude'] = None
df_dados['pb_macro'] = None

# Preencher as colunas no DataFrame de dados com base na correspondência de 'co_municipio_ibge'
for index, row in df_dados.iterrows():
    # Buscar o código do município no DataFrame de população
    cod_mun = row['co_municipio_ibge']
    match = df_populacao[df_populacao['COD_MUN'] == cod_mun]
    
    if not match.empty:
        df_dados.at[index, 'pb_grs'] = match['GRS'].values[0]
        df_dados.at[index, 'pb_regiao_saude'] = match['RS'].values[0]
        df_dados.at[index, 'pb_macro'] = match['Macro'].values[0]
        
# Converter a coluna para o formato datetime e ajustar o formato para dia/mês/ano
df_dados['dt_ocorrencia'] = pd.to_datetime(df_dados['dt_ocorrencia'], errors='coerce')
# Converter a coluna para o formato de string no padrão dia/mês/ano
df_dados['dt_ocorrencia'] = df_dados['dt_ocorrencia'].dt.strftime('%d/%m/%Y')

# Converter a coluna para o formato datetime e ajustar o formato para dia/mês/ano
df_dados['dt_obito'] = pd.to_datetime(df_dados['dt_obito'], errors='coerce')
# Converter a coluna para o formato de string no padrão dia/mês/ano
df_dados['dt_obito'] = df_dados['dt_obito'].dt.strftime('%d/%m/%Y')

# Converter a coluna para o formato datetime e ajustar o formato para dia/mês/ano
df_dados['dt_atestado'] = pd.to_datetime(df_dados['dt_atestado'], errors='coerce')
# Converter a coluna para o formato de string no padrão dia/mês/ano
df_dados['dt_atestado'] = df_dados['dt_atestado'].dt.strftime('%d/%m/%Y')

# Converter a coluna para o formato datetime e ajustar o formato para dia/mês/ano
df_dados['dt_nascimento'] = pd.to_datetime(df_dados['dt_nascimento'], errors='coerce')
# Converter a coluna para o formato de string no padrão dia/mês/ano
df_dados['dt_nascimento'] = df_dados['dt_nascimento'].dt.strftime('%d/%m/%Y')

df_dados['dt_duplicidade'] = pd.to_datetime(df_dados['dt_duplicidade'], errors='coerce')
df_dados['dt_duplicidade'] = df_dados['dt_duplicidade'].dt.strftime('%d/%m/%Y')

df_dados['dt_investigacao'] = pd.to_datetime(df_dados['dt_investigacao'], errors='coerce')
df_dados['dt_investigacao'] = df_dados['dt_investigacao'].dt.strftime('%d/%m/%Y')

df_dados['dt_registro'] = pd.to_datetime(df_dados['dt_registro'], errors='coerce')
df_dados['dt_registro'] = df_dados['dt_registro'].dt.strftime('%d/%m/%Y')


df_dados['dt_ocorrencia'] = pd.to_datetime(df_dados['dt_ocorrencia'], errors='coerce')
df_dados['dt_ocorrencia'] = df_dados['dt_ocorrencia'].dt.strftime('%d/%m/%Y')

df_dados['dt_ultima_atualizacao'] = pd.to_datetime(df_dados['dt_ultima_atualizacao'], errors='coerce')
df_dados['dt_ultima_atualizacao'] = df_dados['dt_ultima_atualizacao'].dt.strftime('%d/%m/%Y')

df_dados['co_municipio_ibge'] = pd.to_numeric(df_dados['co_municipio_ibge'], errors='coerce').fillna(0).astype(int)

df_dados['hr_obito'] = pd.to_datetime(df_dados['hr_obito'], format='%H:%M:%S', errors='coerce').dt.time

df_dados['nu_idade_mae'] = pd.to_numeric(df_dados['nu_idade_mae'], errors='coerce').fillna(0).astype(int)

df_dados['qt_nascidos_vivos'] = pd.to_numeric(df_dados['qt_nascidos_vivos'], errors='coerce').fillna(0).astype(int)

df_dados['co_tipo_ocorrencia'] = pd.to_numeric(df_dados['co_tipo_ocorrencia'], errors='coerce').fillna(0).astype(int)

# Verificar se a conversão foi bem-sucedida
print(df_dados[['dt_ocorrencia', 'dt_duplicidade', 'dt_investigacao', 'dt_registro']].head())

# Verificar o tipo de formato da coluna
for coluna in df_dados.columns:
    print(f"Coluna: {coluna}, Tipo de dados: {df_dados[coluna].dtypes}")
    
# Salvar o DataFrame atualizado em um novo arquivo CSV com encoding UTF-8
output_file_path = r"C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Dev\11-dados_atualizados_com_grs_rs_macro.csv"
try:
    df_dados.to_csv(output_file_path, index=False, encoding='utf-8')
    print(f"Arquivo CSV modificado salvo em: {output_file_path}")
except Exception as e:
    print(f"Erro ao salvar o arquivo CSV: {e}")

# Verificar o DataFrame atualizado
#print(df_dados.head(20))  # Exibir os primeiros 20 registros para verificar
```

```python
# 12 - Tratamento da coluna fx_etaria

#Carregamento do CSV:

#O arquivo CSV é carregado com as colunas de datas dt_nascimento e dt_obito já formatadas como datas (parse_dates=['dt_nascimento', 'dt_obito']).
#Função calcular_faixa_etaria:

#Calcula a idade de uma pessoa com base na diferença em dias entre a data de óbito e a data de nascimento.
#A faixa etária é determinada com base na idade em intervalos de 10 anos (1-10, 11-20, 21-30, etc.).
#Aplicação da Função:

#A função apply é usada para calcular a faixa etária de cada linha, preenchendo a coluna fx_etaria.
#Exportação do CSV:

#O DataFrame atualizado é salvo em um novo arquivo CSV contendo a coluna fx_etaria preenchida com as faixas etárias.
#Este código faz a classificação da faixa etária conforme o intervalo solicitado, considerando as idades dos pacientes no momento do óbito.

import pandas as pd
import numpy as np

# Caminho do arquivo CSV de dados
csv_dados_path = r"C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Dev\11-dados_atualizados_com_grs_rs_macro.csv"

# Carregar o arquivo CSV de dados
try:
    df = pd.read_csv(csv_dados_path, delimiter=',', encoding='utf-8', parse_dates=['dt_nascimento', 'dt_obito'], dayfirst=True)
    print("Arquivo CSV carregado com sucesso.")
except Exception as e:
    print(f"Erro ao carregar o arquivo CSV: {e}")
    exit()

# Função para calcular a faixa etária com base na idade
def calcular_faixa_etaria(dt_nascimento, dt_obito):
    if pd.isnull(dt_nascimento) or pd.isnull(dt_obito):
        return np.nan
    idade = (dt_obito - dt_nascimento).days // 365
    faixa_inicio = (idade // 10) * 10 + 1
    faixa_fim = faixa_inicio + 9
    return f"{faixa_inicio}-{faixa_fim}"

# Aplicar a função para calcular e preencher a coluna 'fx_etaria'
df['fx_etaria'] = df.apply(lambda row: calcular_faixa_etaria(row['dt_nascimento'], row['dt_obito']), axis=1)

# Verificar as primeiras linhas do DataFrame atualizado
print(df[['dt_nascimento', 'dt_obito', 'fx_etaria']].head(20))

# Salvar o DataFrame atualizado em um novo arquivo CSV
output_file_path = r"C:\Users\anton\Desktop\ReapQuali\Obitos_Python\Teste\12-dados_atualizados_com_fx_etaria.csv"
try:
    df.to_csv(output_file_path, index=False, encoding='utf-8')
    print(f"Arquivo CSV modificado salvo em: {output_file_path}")
except Exception as e:
    print(f"Erro ao salvar o arquivo CSV: {e}")
```
